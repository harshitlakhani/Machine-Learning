```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import image

import torch
import cv2
from torch.nn import Sequential 
from torch.nn import MaxPool2d, AvgPool2d  
```

```python
path = "/content/Triangle_1ef8e34c-9cab-11ec-b61e-235a07e81dcd.png"
figure_size = (8,8)
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=figure_size)
plt.imshow(img)
plt.title("Original Image")
plt.show()
```

```python
img = torch.from_numpy(img)
img = img.permute(2,0,1)
img = img.unsqueeze(dim=0)
img.shape
```

```python
KERNEL_SIZE=2
STRIDE=2
model = MaxPool2d(kernel_size = KERNEL_SIZE, stride = STRIDE)
  
# generate pooled output 
output = model(img.float()) 
  
output = output.squeeze()
plt.figure(figsize=figure_size)
plt.imshow(output.permute(1,2,0).int())
plt.title("MaxPooling2D with kernel_size = {}, stride = {}".format(KERNEL_SIZE, STRIDE))
plt.show()
```

```python
model = AvgPool2d(kernel_size = KERNEL_SIZE, stride = STRIDE)
  
# generate pooled output 
output = model(img.float()) 

output = output.squeeze()
plt.figure(figsize=figure_size)
plt.imshow(output.permute(1,2,0).int())
plt.title("AvgPooling2D with kernel_size = {}, stride = {}".format(KERNEL_SIZE, STRIDE))
plt.show()
```

Here the black and white images are of RGB channels but all the channels have same pixel values so, ultimately read it as a RGB image doesn't make any differences.
